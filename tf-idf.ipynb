```python
import math
dictonary = vocab = {
    "hi": 0,
    "hello": 1,
    "hey": 2,
    "friend": 3,
    "buddy": 4,
    "mate": 5,
    "colleague": 6,
    "partner": 7
}
D1= "hi hello hey"
D2= "hello friend buddy"
D3= "mate friend colleague partner"
D4= "hi friend partner"
def vectortizer(doc,dictonary):
    d = doc.split()
    vector = []
    for i in dictonary:
        vector.append(d.count(i)/len(d))
    return vector

d = [D1,D2,D3,D4]
DF = 0
N = len(d)
idf = []
for i in dictonary:
    DF = 0
    for j in d:
        if i in j.split():
            DF += 1
    idf.append(math.log(N/(DF+1))+1)
    
```

```python
def dot_product(vector1,vector2):
    result = 0
    for i in range(len(vector1)):
        result += vector1[i][1]*vector2[i][1]
    return result
```

```python
def magnitude(vector):
    sum_of_squares = 0
    for val in vector:
        item = val[1]
        sum_of_squares += item * item  # Square the value
    return math.sqrt(sum_of_squares)
```

```python
def cosine_similarity(vector_a, vector_b):
    dot = dot_product(vector_a, vector_b)
    mag_a = magnitude(vector_a)
    mag_b = magnitude(vector_b)
    
    # Safety check: If a vector is all zeros (length 0), we can't divide
    if mag_a == 0 or mag_b == 0:
        return 0
        
    return dot / (mag_a * mag_b)
```

```python
def vec1(di):    
    tf_idfvector = []
    vector = vectortizer(di,dictonary)
    for i in range(len(dictonary)):
        tf_idfvector.append([i,idf[i]*vector[i]])
    return tf_idfvector
```

```python
D5 = input().lower()
vector_a = vec1(D5)
res  = [] 
A = [D1,D2,D3,D4]
for i in A:
    res.append([i,cosine_similarity(vector_a,vec1(i))])
print(max(res,key=lambda x :x[1]))
```
